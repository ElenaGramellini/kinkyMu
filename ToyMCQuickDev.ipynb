```python
# we are trying to reproduce the muon cosmic value from the paper
# step 1: write down the equation 
# step 2: import some data at sea level for comparison (and theta = 0)
# check: plot one on top of the other
import numpy as np
import matplotlib.pyplot as plt

# example parametric function
def flux(E, theta, Ec):
    I0 = 88.0
    E0 = 4.29 # GeV
    epsilon = 854
    n = 3.09
    Rd = 174
    D = np.sqrt(Rd*Rd*np.cos(theta) + 2*Rd + 1 ) - Rd*np.cos(theta)
    N = (n - 1)*((E0 + Ec)**(n - 1))
    N /= 10000
    P0 = I0*N
    P1 = (E0 + E)** (-n)
    P2 = (1+E/epsilon)**(-1)
    P3 = D**(-n+1)
    x = P0*P1*P2*P3
    return x


# Generate parameter values
t_values = np.linspace(0, 10000, 10000)

# Evaluate the parametric function
y_values = flux(t_values, 0, 10000)

# Create the plot
plt.loglog(t_values, y_values)
#plt.loglog(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Parametric Function: Flux')
plt.grid(True)

# Show the plot
plt.show()
```

```python
#example of 2D pdf normalized to 1 
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the Gaussian distribution
mu = np.array([0, 0])  # Mean vector
cov = np.array([[1, 0], [0, 1]])  # Covariance matrix

# Generate the grid of x, y values
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Calculate the PDF values for each (x, y) point
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y
Z = (1 / (2 * np.pi * np.linalg.det(cov) ** 0.5)) * np.exp(-0.5 * np.sum((pos - mu) @ np.linalg.inv(cov) * (pos - mu), axis=2))

# Normalize the PDF to one
Z_normalized = Z / np.sum(Z)

# Create a contour plot of the PDF
plt.contourf(X, Y, Z_normalized)
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D PDF Normalized to One')

# Display the plot
plt.show()
```

```python
#make a box in the center of the coordinates
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and a 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the dimensions of the box
width = 7
height = 6
depth = 7

# Compute the corner points of the box
x = [-width/2, -width/2, width/2, width/2, -width/2, -width/2, width/2, width/2]
y = [-height/2, height/2, height/2, -height/2, -height/2, height/2, height/2, -height/2]
z = [-depth/2, -depth/2, -depth/2, -depth/2, depth/2, depth/2, depth/2, depth/2]

# Plot the box
ax.plot(x, y, z)

# Set the limits and labels for each axis
ax.set_xlim([-width/2, width/2])
ax.set_ylim([-height/2, height/2])
ax.set_zlim([-depth/2, depth/2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the title
plt.title('3D Box')

# Display the plot
plt.show()
```
